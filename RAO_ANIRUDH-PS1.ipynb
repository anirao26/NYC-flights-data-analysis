## Question 4
Which flights (i.e. carrier + flight + dest) happen every day? Where do they fly to?

```python
# Your code here
flight_count_carrier_flight_dest = flights_df.groupby(['carrier', 'flight', 'dest']).size()
everyday_flights = flight_count_carrier_flight_dest[flight_count_carrier_flight_dest/365==1]
everyday_flights = everyday_flights.index.values.tolist()

destinations = []
for flight in everyday_flights:
    destination = flight[2]
    if(destination not in destinations):
        destinations.append(destination)

print("Everyday flights are:")
for flight in everyday_flights:
    print(flight[0], flight[1], flight[2])
    
print("\nDestinations flown to by the flights are:")
for destination in destinations:
    print(destination)
```

The above combinations of carrier, flight and dest happen everyday. The dest attribute tells us the destinations they fly to.

## Question 6
What weather conditions are associated with flight delays leaving NYC? Use graphics to explore.

```python
weather_data = pd.read_csv("weather.csv")
weather_df = weather_data.groupby(['year','month','day','origin']).mean()
weather_df.reset_index(level = 0, inplace = True)
weather_df.reset_index(level = 0, inplace = True)
weather_df.reset_index(level = 0, inplace = True)
weather_df.reset_index(level = 0, inplace = True)

flights_agg_mean = flights_df.groupby(['year','month','day','origin']).mean()
flights_agg_mean.reset_index(level = 0, inplace = True)
flights_agg_mean.reset_index(level = 0, inplace = True)
flights_agg_mean.reset_index(level = 0, inplace = True)
flights_agg_mean.reset_index(level = 0, inplace = True)

flights_weather_merge = pd.merge(flights_agg_mean, weather_df, on=['year','month','day','origin'], how='inner')

import seaborn as sns
%matplotlib inline

sns.pairplot(flights_weather_merge, x_vars = ['temp','dewp','humid'], y_vars = 'dep_delay', size = 7,
             aspect = 0.7, kind='reg')
sns.pairplot(flights_weather_merge, x_vars = ['wind_dir', 'wind_speed','wind_gust'], y_vars = 'dep_delay', size = 7,
             aspect = 0.7, kind='reg')
sns.pairplot(flights_weather_merge, x_vars = ['precip', 'pressure','visib'], y_vars = 'dep_delay', size = 7,
             aspect = 0.7, kind='reg')
```


After examining the scatter-plots of departure delays with respect to all the weather factors the following conclusions can be drawn:

1. There seems to be a fairly strong positive correlation between humidity and departure delay.
2. There seems to be a strong positive correlation between precipitation and departure delay.
3. There seems to be a fairly weak positive correlation between pressure and departure delay.
4. There seems to be a strong negative correlation between visibility and departure delay.

```python
from sklearn.linear_model import LinearRegression
x = flights_weather_merge[['temp','precip','humid','visib']]
y = flights_weather_merge['dep_delay']
linreg = LinearRegression()
linreg.fit(x, y)
print (linreg.intercept_)
print (linreg.coef_)
list(zip(['temp','precip','humid','visib'], linreg.coef_))
```

The equation of the linear regression model is:

Departure Delay (y) = 0.05 x temperature + 1.83 x precipitation + 1.94 x humidity - 1.8 x visibility + 16.18

The slope values for each of the predictors represent the change in departure delay if the predictor variable is increased by 1 unit. The intercept is 16.18 units.
